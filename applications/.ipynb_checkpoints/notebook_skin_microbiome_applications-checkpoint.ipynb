```r
# load package and set global seed
library(BayesTime)
library(gridExtra)
library(cowplot) 
library(gridGraphics) 
set.seed(31)
```

```r
# read in longitudinal dataset
df = read.csv("data_skin_microbiome.txt", sep='\t')
head(df)
```

```r
## prepare data for SFPCA
dat <- prepare_data(data = df, unique_subject_id = 'ID_unique', time_name = 'timepoint',
                    response_name = 'shannon_avg', transform_y = 'standardize', scale_time = T)
```

```r
# sfpca_stan_results <- stan_fit(sfpca_data = dat, Nsamples = 1000, Nchain = 3, Ncores=3, 
#                                PC_range = c(2,3,4), nknot_range = c(2,3,4))
# save(sfpca_stan_results, file='bayesTime_modelSelection.RData') 
```

```r
load('../../bayesTime_modelSelection.RData')
```

```r
# choose optimal model 
optimal_model_idx <- optimal(model_list = sfpca_stan_results)
```

```r
# use default model with highest elpd from PSIS-LOO
optimal_model <- sfpca_stan_results[[optimal_model_idx]]
model_output <- output_results(sfpca_data = dat, model = optimal_model)
```

### Figure 1

```r
# check model fit
fig_posterior = plot_posterior_diagnostic(dat, optimal_model)$figure 
fig_posterior
```

```r
# identify outliers
fig_looic = plot_k_diagnostic(dat, optimal_model)$figure
fig_looic
```

```r
# examine outliers
fig_outlier1 = plot_obs_predict(sfpca_data=dat, optimal_model=optimal_model, data=df, 
                 time_name = 'timepoint', response_name = 'shannon_avg', 
                 unique_subject_id = 'ID_unique', subject_selected='Volunteer 9Armpit',
                 x_lab='Weeks', y_lab='Shannon diversity', 
                 title='Outlier 46: observed vs. fitted', legend_show=FALSE,
                 x_axis_ticks=seq(0,8,2), x_axis_range=c(0,8), 
                 y_axis_ticks=seq(0,8,2), y_axis_range=c(0,8))$figure
fig_outlier1
```

```r
fig_outlier2 =  plot_obs_predict(sfpca_data=dat, optimal_model=optimal_model, data=df, 
                 time_name = 'timepoint', response_name = 'shannon_avg', 
                 unique_subject_id = 'ID_unique', subject_selected='Volunteer 6Foot',
                 x_lab='Weeks', y_lab='Shannon diversity', 
                 title='Outlier 36: observed vs. fitted', legend_show=FALSE,
                 x_axis_ticks=seq(0,8,2), x_axis_range=c(0,8), 
                 y_axis_ticks=seq(0,8,2), y_axis_range=c(0,8))$figure
fig_outlier2
```

```r
plot_grid(fig_posterior, fig_looic, 
          fig_outlier1 , 
          fig_outlier2 ,
          labels = c('A', 'B', 'C', 'D'), 
          hjust = 0, vjust = 1, ncol=2) 
ggsave('application_fig1.png', width=8, height=8, dpi=300)
```

### figure 2

```r
fig_mean = plot_mean_curve(output = model_output, original=T, 
                ymin=0.5, ymax=8, x_lab='Weeks', y_lab='Shannon diversity')$figure
```

```r
fig_fpc = plot_fpc_curve(output = model_output, pc_idx=c(1,2,3,4), ymin=2, ymax=7.5,
               original=T, x_lab='Weeks', y_lab='Shannon diversity')$figure
fig_fpc
```

```r
fig_pc1 = plot_fpc_on_mean_curve(output = model_output, pc_idx=1, original = T, sd=T, 
                       x_lab='Weeks', y_lab='Shannon diversity', ymin=0.5, ymax=8)$figure
fig_pc1
```

```r
fig_pc2 = plot_fpc_on_mean_curve(output = model_output, pc_idx=2, original = T, sd=T,
                       x_lab='Weeks', y_lab='Shannon diversity', ymin=0.5, ymax=8)$figure
fig_pc2
```

```r
fig_score_pc1 = plot_fpc_boxplot(output = model_output, pc_idx=1, group_name = 'sample_site',
                 x_lab = 'Body site', y_lab = 'Score of the first PC', p_title='Shannon diversity',
                 testing_type = 'non-parametric', pairwise_testing=F)$figure 
```

```r
fig_score_pc2 = plot_fpc_boxplot(output = model_output, pc_idx=2, group_name = 'sample_site',
                 x_lab = 'Body site', y_lab = 'Score of the second PC', p_title='Shannon diversity',
                 testing_type = 'non-parametric', pairwise_testing=F, pval_show_all=T)$figure 
```

```r
plot_grid(fig_mean, fig_fpc, 
          fig_pc1, fig_pc2,
          fig_score_pc1, fig_score_pc2,
          labels = c('A', 'B', 'C', 'D', 'E', 'F'), 
          hjust = 0, vjust = 1, ncol=2) 
ggsave('application_fig2.png', width=7, height=10, dpi=300)
```
